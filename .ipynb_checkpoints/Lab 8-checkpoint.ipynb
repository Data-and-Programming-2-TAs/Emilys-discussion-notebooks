## A Quick Note on ML & Algorithmic Bias
Machine learning, like most of the topics you’re learning this quarter, is a very deep discipline that we’re only going to touch the surface of. Unlike most other disciplines you’re learning this quarter, you can do quite a bit of damage if you apply machine learning incorrectly. It is of the utmost importance to apply machine learning carefully and always with an eye towards fairness, accountability, and explainability.Algorithms themselves are not biased, but they can return results that are biased against certain groups of people if the data fed into them is biased. The data is almost guaranteed to be biased if you are using data from the real world, because the real world is biased against certain groups of people. For example, the National Institute of Standards and Technology recently found that facial recognition algorithms are 10 to 100 times more likely to inaccurately identify an East Asian or black face compared to a white face (<a href=“https://www.scientificamerican.com/article/how-nist-tested-facial-recognition-algorithms-for-racial-bias”>Scientific American</a>).This is even more of a problem when algorithms determine important decisions, like whether an individual is <a href=“https://www.law.upenn.edu/live/files/7952-kearns-finalpdf”>eligible for a loan</a>, or <a href=“https://www.businessinsider.com/amazon-ai-biased-against-women-no-surprise-sandra-wachter-2018-10”>whether to invite an individual for a job interview</a>. There is a developing literature on this topic, known as FATML (Fairness, Accountability, and Transparency in Machine Learning). There are no one size fits all solutions, but there are some basic steps you can take, like evaluating the outcomes based on all types of groupings of individuals, using tools like <a href=“https://github.com/dssg/aequitas”>aequitas</a> (developed by some Harris folks among others), and generally being sensitive and attentive to the potential biases your algorithm might propagate.

### Train-Test Split

One of the key steps in supervised ML is to split the data into training and test sets. (This is what distinguishes Linear Regression in stats from Linear Regression in the ML world). The idea is that we'll train the model on the training set (often using a grid search of many different models and hyperparameters), then test it using the test set. During the testing step, we use metrics to evaluate the model fit.

We might also want to split out a validation step, which is used for something called cross-validation.




```python
X = wine_df.loc[:, wine_df.columns != 'quality']
y = wine_df['quality']

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

### Imputing data
We'll want to deal with missing values. A common way to do this is to impute missing values with the median or mean, for all continuous variables.

```python
## Note: you can use sklearn's imputation function, which the professor showed in class:
# from sklearn.impute import SimpleImputer
# imputer = SimpleImputer(strategy='mean')
# imputed_data = imputer.fit_transform(df)
# imputed_data

# OR - you can write your own code. Sometimes, you may not want the format/steps that sklearn 
# provides, e.g. sklearn usually outputs data in an array rather than a dataframe, so you lose the column information.

def impute_missing(X_train, X_test, cols):
    for df in [X_train, X_test]:
        for i in cols:
            median_col = df[i].median()
            df[i] = df[i].fillna(median_col)
    return X_train, X_test
```

```python
X_train, X_test = impute_missing(X_train, X_test, X_train.columns)
```

## Model selection
### Supervised vs. Unsupervised learning

**Supervised:** When we have a labeled outcome variable (e.g. quality)
- Regression: predicting a continuous quantity
- Classification: predicting a discreete class label


**Unsupervised:** don't have a labeled outcome variable - we're just trying to find patterns in the data



### Grid search
A common way to run through machine learning models is to create a grid search. In a grid search, you'll want to specify which models to run, and the hyperparameters. A lot of times, you'll want to run a bunch of models for the appropriate data type. Hyperparameters are parameters that you can specify for each model.

The example below is a CLASSIFICATION grid search, and we're running logistic regression, support vector machine, and naive bayes. 

If we were to run this as a REGRESSION problem, we might include linear regression (incl. ridge and lasso), elastic net, decision tree, random forest, etc.

#### Steps:
1. Fit each of the models using the TRAINING data --> model.fit(X_train, y_train)
2. Use the fit model to predict the TEST data --> Y_hat = predict(X_test)
3. Then, you use metrics to determine how close your predictions are to the true y values --> compare y_hat to y_train

### Evaluation Metrics for Supervised Learning
- Mean Squared Error
- Mean Absolute Error
- Accuracy (Classification)
- Precision (Classification)
- Recall (Classification)

```python
# Loop over models 
def grid_search(MODELS, GRID, X_train, y_train, X_test, y_test):
        
    # Begin timer 
    start = datetime.datetime.now()
    
    # Initialize results data frame 
    results_df = pd.DataFrame()
    #results_df = pd.DataFrame(columns=['Model type', 'C', 'Penalty', 'Accuracy'])

    for model_key in MODELS.keys(): 
        
        # Loop over parameters 
        for params in GRID[model_key]: 
            print("Training model:", model_key, "|", params)
            
            # Create model 
            model = MODELS[model_key]
            model.set_params(**params)
            
            # Fit model on training set 
            fit_model = model.fit(X_train, y_train.values.ravel())
            
            # Predict on testing set 
            y_pred = fit_model.predict(X_test)
            y_true = y_test
            
            # Evaluate predictions 
            accuracy = accuracy_score(y_true, y_pred)
            mse = mean_squared_error(y_true, y_pred)
            mae = mean_absolute_error(y_true, y_pred)
            #r2 = r2_score(y_true, y_pred)
            
            # Store results in your results data frame
            row = params
            row['model'] = model_key
            row['accuracy'] = accuracy
            row['mean squared error'] = mse
            row['mean absolute error'] = mae
           # row['r-squared'] = r2
            if 'random_state' in X_train.columns:
                del row['random_state']
            results_df = results_df.append(row, ignore_index=True)

            
        # End timer
        stop = datetime.datetime.now()
        print("Time Elapsed:", stop - start) 
    return results_df
    
```

```python
# Config: Dictionaries of models and hyperparameters
MODELS = {
    'LogisticRegression': LogisticRegression(), 
    'LinearSVC': LinearSVC(), 
    'GaussianNB': GaussianNB()
}

GRID = {
    'LogisticRegression': [{'penalty': x, 'C': y, 'random_state': 0, 'max_iter': 5000} 
                           for x in ('l2', 'none') \
                           for y in (0.01, 0.1, 1, 10, 100)],
    'GaussianNB': [{'priors': None}],
    'LinearSVC': [{'C': x, 'random_state': 0, 'max_iter': 2000} \
                  for x in (0.01, 0.1, 1, 10, 100)]
}

results_df = grid_search(MODELS, GRID, X_train, y_train, X_test, y_test)
```
